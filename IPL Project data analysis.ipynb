```python
# IPL matches played by each team
plt.figure(figsize=(10,8))

data = df['bowling_team'].value_counts().sort_values(ascending=False)

sns.barplot(y=data.index, x=data, orient='h')   # horizontal bar chart
plt.xlabel('Matches Played')
plt.ylabel('Team')
plt.show()
```

```python
# Most runs scored by IPL teams
df.groupby('batting_team')['run'] .sum() .sort_values(ascending=False)
```

```python
# Most IPL runs by a batsman
plt.figure(figsize=(10,8))

data = (
    df.groupby('striker')['runs_off_bat'].sum().sort_values(ascending=False).head(10))
sns.barplot(y=data.index, x=data, orient='h')
plt.xlabel('Batsman')
plt.ylabel('Runs')
plt.show()
```

```python
#Avg Run by Teams in Powerplay
df[df['over'] < 6].groupby(['match_id', 'batting_team'])['run'].sum() .groupby('batting_team').mean().sort_values(ascending=False)
```

```python
#most ipl century by a player
runs = df.groupby(['striker','match_id'])['runs_off_bat'].sum()
runs[runs >= 100].droplevel(1).groupby('striker').count().sort_values(ascending=True)
```

```python
# Most IPL fifty by player
plt.figure(figsize=(10,8))

runs = df.groupby('striker').count().sort_values(by='run', ascending=False)[:10]

sns.barplot(y=runs.index, x=runs['run'], orient='h')
plt.xlabel('Half-centuries')
plt.ylabel('batsman')
plt.show()
```

```python
df[df['runs_off_bat'] == 6] .groupby(['start_date','striker']) .count()['season'].sort_values(ascending=False)[:10]
```

```python
# most boundary (4s) hit by a batsman
plt.figure(figsize=(10,8))
data = df[df['runs_off_bat'] == 4].groupby('striker')['runs_off_bat'].count().sort_values(ascending=False)[:10]

data.plot(kind='bar')
```

```python
#most runs in on ipl season by player
df.groupby(['striker','season'])['runs_off_bat'].sum().sort_values(ascending=False)[:10]
```

```python
#no of sixes in ipl seasons
plt.figure(figsize=(10,8))
data = df[df['runs_off_bat'] == 6].groupby('season')['match_id'].count().sort_values(ascending=False)

sns.barplot(y=data.index, x=data.values, orient='h')

plt.xlabel('sixes')
plt.ylabel('season')
plt.show()
```

```python
# highest total by IPL teams
df['total'] = df['runs_off_bat'] + df['extras']
df.groupby(['start_date','batting_team'])['total'].sum().sort_values(ascending=False)[:10]
```

```python
#most ipl sixes hit by a batsman
plt.figure(figsize=(10,8))
data= df[df['runs_off_bat']==6]['striker'].value_counts()[:10]
sns.barplot(y=data.index,x=data,orient='h')
plt.xlabel('sixes')
plt.ylabel('batsman')
plt.show()
```

```python
# highest individual IPL score
df.groupby(['striker','start_date'])['runs_off_bat'].sum().sort_values(ascending=False)[:10]
```

```python
#most run conceded by a bowler in on inning
df.groupby(['bowler','start_date'])['run'].sum().droplevel(level=1).sort_values(ascending=False)[:10]
```

```python
# most dot balls by a bowler
plt.figure(figsize=(10,8))

data = df[df['runs_off_bat'] == 0] \
        .groupby('bowler')['match_id'] \
        .count() \
        .sort_values(ascending=False)[:10]

sns.barplot(y=data.index, x=data.values, orient='h')

plt.xlabel('dot balls')
plt.ylabel('bowler')
plt.show()
```

```python
# most IPL wickets by a bowler
wicket_list = ['caught','bowled','lbw','stumped','caught and bowled','hit wicket']

data = df[df['wicket_type'].apply(lambda x: x in wicket_list)]['bowler'].value_counts()[:10]
data
```

```python
# Most maiden overs by a bowler
data = df.groupby(['start_date', 'bowler', 'over'])['runs_off_bat'].sum()

data = data[data == 0]           # only maiden overs
data = data.droplevel([0, 2])    # remove start_date & over

data.value_counts()[:10]
```

```python
# most wickets by an IPL team
plt.figure(figsize=(10,8))

wicket_list = ['caught','bowled','lbw','stumped','caught and bowled','hit wicket']

data = df[df['wicket_type'].apply(lambda x: x in wicket_list)]['bowling_team'].value_counts()

sns.barplot(y=data.index, x=data.values, orient='h')

plt.xlabel('wickets')
plt.ylabel('Teams')
plt.show()
```
